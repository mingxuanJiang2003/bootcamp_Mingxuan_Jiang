```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

data_path = Path("../data/market_daily_sample.csv")
df = pd.read_csv(data_path, parse_dates=["date"]) 
df = df.sort_values("date")
df.head()
```

```python
plt.figure()
plt.plot(df['date'], df['spy_close'])
plt.title('SPY Close (Synthetic)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure()
plt.plot(df['date'], df['vix_close'])
plt.title('VIX (Synthetic)')
plt.xlabel('Date')
plt.ylabel('Level')
plt.show()
```
